```python
## clustering using e-m clustering method
import ast
import requests

link = "https://bradfordtuckfield.com/emdata.txt"
f= requests.get(link)
allpoints= ast.literal_eval(f.text)

print(allpoints[0:20])
```

```python
## now getting only two clusters
allxs= [point[0] for point in allpoints]
allys= [point[1] for point in allpoints]
print(allxs[0:10])
```

```python
## now plotting them we get
import matplotlib.pyplot as plt
plt.plot(allxs,allys,'x')
plt.axis('equal')
plt.show()
```

```python
## now we have to classify them how do we do is by using a classification code ## em clustering step 1 Guessing the means
from scipy.stats import multivariate_normal
mean1=[-1,0]
mean2=[0.5,-1]
mean3=[0.5,0.5]

allmeans= [mean1,mean2,mean3]

cov1=[[1,0],[0,1]]
cov2=[[1,0],[0,1]]
cov3=[[1,0],[0,1]]

allvar= [cov1,cov2,cov3]

## this is e-m clustering step 2 Expectation classifyin based on the guessed mean by comparing the bell curves

def classify(allpts,allmns,allvar):
    vars=[]
    for n in range(len(allmns)):
        vars.append(multivariate_normal(mean=allmns[n], cov=allvar[n]))
    classification=[]
    for point in allpts:
        this_classification=-1           ## an important prog point here is we took these variables here in for loop insteead
        this_pdf=0                    ##of out of the loop to again start it with -1 and 0 initialization whenever the new point
        for n in range(len(allmns)): ## changes to avoid comparing hurdles and wrong classification
            if vars[n].pdf(point)>this_pdf:
                this_pdf=vars[n].pdf(point)
                this_classification=n+1
        classification.append(this_classification)
    return classification
```

```python
theclass=classify(allpoints,allmeans,allvar)
print(theclass[0:20])
```

```python
print(len(allmeans))
print(len(allvar))
```

```python
## now plotting the clusters with the guessed mean as their clusters using a function
def makeplot(allpts,theclass,allmns):
    thecolors= ['black']*len(allpts)
    for idx in range(len(allpts)):
        if theclass[idx]== 2:
            thecolors[idx]= 'green'
        if theclass[idx]== 3:
            thecolors[idx]= 'yellow'
    allxs= [point[0] for point in allpts]
    allys= [point[1] for point in allpts]
    for i in range(len(allpts)):
        plt.plot(allxs,allys,thecolors[i])
    for i in range(len(allmns)):
        plt.plot(allmns[i][0],allmns[i][1],'x')
    plt.axis('equal')
    plt.show()
    
makeplot(allpoints,theclass,allmeans)        
    
```

```python
## Maximization step we get the correct centers that is the correct means and variances
import numpy as np

def getcenters(allpoints,theclass,k):
    centers=[]
    thevars=[]
    for n in range(k):
        pointsn=[allpoints[i] for i in range(0,len(allpoints)) if theclass[i]==(n+1)]  ## theclass is used to access the simultaneous positions of the classified clusters or points(x,y)
        xpointsn= [points[0] for points in pointsn]
        ypointsn=[points[1] for points in pointsn]
        xcenter=np.mean(xpointsn) ## remember that we have to take the mean of every cluster each and every one.
        ycenter=np.mean(ypointsn)
        centers.append([xcenter,ycenter])
        thevars.append(np.cov(xpointsn,ypointsn))
    return centers,thevars
```

```python
allmeans,allvars= getcenters(allpoints,theclass,3) ## get the calculated means and variances of the 3 clusters using 
## expected classification
```

```python
theclass= classify(allpoints,allmeans,allvars)## as we have calculated new means we must get the correct classifications so 
## we do the classification again as it was also expected not accurate
```

```python
makeplot(allpoints,theclass,allmeans) ## but still not in the center
```

```python
allmeans,allvars= getcenters(allpoints,theclass,3)## as we have now got the correct classification based on the calculated'means
## so initially we have taken the expected classification in the function so we perform again with new classification for correc
## t results.
```

```python
## because we got the calculated centers means and variances again we have to calculate the classification this time based on 
## correct centers

for n in range(0,100): ## whenever we do e step (theclass) we have to perform the maximization m (getcenters) for convergence
    theclass= classify(allpoints,allmeans,allvars) ## so we loop them so that they do meet in the center.
    allmeans,allvars= getcenters(allpoints,theclass,3)
```

```python
makeplot(allpoints,theclass,allmeans) ## now plot the final convergence (perfect)
```

```python
## item based collaborative filtering 
## helps us recommend those items which have a higher similarity of getting purchased with a specific product already chosen
## the thing here is that first a product must already be chosen then we can recommend based on similarity of purchase
import numpy as np

def dotproduct(vector1,vector2):
    thedotproduct= np.sum([vector1[k]*vector2[k] for k in range(len(vector1))])
    return(thedotproduct)

## now we need the norm of the vectors as in the cos formula
def vectornorm(vector):
    thenorm= np.sqrt(dotproduct(vector,vector))
    return(thenorm)

def cosinesimilarity(vector1,vector2):
    thesimilarity= (dotproduct(vector1,vector2)/vectornorm(vector1)*vectornorm(vector2))
    thesimilarity= np.round(thesimilarity,4)
    return(thesimilarity)
```

```python
##now using these functions onn any type of item data similarity we want we can take it 
## suppose if the user or the customer has purchased item1 and you have to recommend a item to him which is mostly purchased
## with the selectem item1 so we use the cosinesimilarity function as follows
similarity= []
ouritem= 'item1'
recommenditem= [items for items in interaction.index if items!= ouritem] ## this means that first recommenditem itself becomes
## a complete list and then goes for the next step therecommendeditem , thats why we are getting a complete table and not 
## specific value

therecommendeditem= interaction.loc[recommenditem,:]

takenitem= interaction.loc[ouritem,:]

similarity.append(cosinesimilarity(takenitem,therecommendeditem))
interaction.loc[:'similarity']= similarity
recommendations= list(interaction.sort_values(by='similarity',ascending= False))
print(recommendations)
```

```python
## function to implement item based colloaborative filtering
def get_item_recommendations(interaction,itemname):
    similarity=[]
    ouritem= itemname
    recommenditem= [items for items in interaction.index if items!= ouritem]
    
    recommendeditem= interaction.loc[recommenditem,:]
    theitem= list(interaction.loc[ouritem,:])
    for item in recommendeditem.index:
        similarity.append(cosinesimilarity(theitem,list(recommendeditem.loc[item,:])))
    recommendeditem.loc[:,'similarity']= similarity
    recommendations= list(recommendeditem.sort_values(by='similarity',ascending= False).index)
    return(recommendations)
```

```python
import pandas as pd
import numpy as np  ## we are taking an interaction matrix like dataset
interaction=pd.read_csv('https://bradfordtuckfield.com/purchasehistory1.csv')## here links can be also used to read
interaction.set_index("Unnamed: 0", inplace = True)
print(interaction)
```

```python
def dotproduct(vector1,vector2):
    thedotproduct= np.sum([vector1[k]*vector2[k] for k in range(len(vector1))])
    return(thedotproduct)

## now we need the norm of the vectors as in the cos formula
def vectornorm(vector):
    thenorm= np.sqrt(dotproduct(vector,vector))
    return(thenorm)

def cosinesimilarity(vector1,vector2):
    thesimilarity= (dotproduct(vector1,vector2)/vectornorm(vector1)*vectornorm(vector2))
    thesimilarity= np.round(thesimilarity,4)
    return(thesimilarity)
```

```python
##user based collaborative filtering
## now we have to find the user most similar to the user who is currently purchasing our product or simply put find the users
## similar to the user we want an item to recommend
def get_similar_users(interaction,username):
    othercolumns=[columnname for columnname in interaction.columns if columnname!=username]
    otherusers=interaction[othercolumns]
    theuser=list(interaction[username])
    similarities=[]
    for users in otherusers.columns:
        similarities.append(cosinesimilarity(theuser,list(otherusers.loc[:,users])))
        otherusers.loc['similarities',:]=similarities
        return list(otherusers.sort_values(by='similarities',axis=1,ascending=False).columns)
```
